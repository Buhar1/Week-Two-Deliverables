```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
path='company_sales_data.csv'

try:
    file=pd.read_csv(path)
    print('File Loaded.')
except Exception as e:
    print(f'An error occurred while loading the file {e}.')
    exit()

    
```

```python
plt.figure(figsize=(6,6))
plt.plot(file['month_number'], file['total_profit'], color='b')

plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.show()
```

```python
fig, axes=plt.subplots(nrows=2, ncols=1)
axes[0].plot(file['month_number'], file['bathingsoap'], color='b')
axes[0].set_ylabel('Total Units')

axes[1].plot(file['month_number'], file['facewash'], color='red')
axes[1].set_ylabel('Total Units')
axes[1].set_xlabel('Month')

plt.show()
```
